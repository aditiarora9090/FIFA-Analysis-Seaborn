# Answer 1

```python
# importing necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

## Insight 1

```python
# Overall rating of the players
sns.distplot(df['overall'], bins=10, color = 'r')
plt.title("Distribution of Overall ratings of all Players")
plt.show()
```

From the above plot, we get to know that the overall rating of the players is normally distributed. 

## Insight 2

```python
# Potential and Wage_EUR, filtered by International Reputation
sns.scatterplot(x =df['potential'], y = df['wage_eur'], hue = df['international_reputation'])
```

From the above graph, we can draw an insight that there is a direct relationship between potential and wages i.e higher the potential, higher the wages in EUR of the people. Also, we can see that people with higher wages also have higher international reputation.

# Insight 3

```python
# Age and Preferred Foot
plt.bar(df['preferred_foot'],df['age'])
```

From the above graph, we can conclude that the players with higher age prefer using their right foot for playing the match.

# Insight 4 

```python
# Players based upon Overall rating
plt.bar(df['short_name'].head(),df['overall'].head())
```

```python
df['short_name'].head()
```

We can say that L. Messi is the best football player with the highest number of overall rating, followed by R. Lewandowski and Christiano Ronaldo.

# Insight 5

```python
# Players and their wages_EUR
rich = df.sort_values('wage_eur', ascending = False)
plt.bar(rich['short_name'].head(),rich['wage_eur'].head())
```

We can say that K. De Bruyne and K. Benzema are the richest players with the highest wages of approximately 350000 EUR.

# Insight 6

```python
# finding the oldest player
oldest = df.sort_values('age', ascending = False)
plt.bar(oldest['short_name'].head(),oldest['age'].head())
```

We can conclude that K. Miura is the oldest player with the age nearly 55 years.

# Insight 7

```python
# Clubs with most number of players
df_clubname = df['club_name'].value_counts()
```

```python
df_clubname
```

```python
df_clubname.sort_values(ascending = False).head(25).plot(kind ='bar')
```

We can conclude that the club names plotted above have the highest demand in the market with the highest number of players.

# Question 2

## (a) Selection of Goalkeeper

```python
df[(df['player_positions'] == 'GK') & (df['club_name'] == 'Real Madrid CF')]
```

#### We have 3 goalkeepers from Real Madrid CF club, but we've to find the best goalkeeper so let's look into the goalkeepers of other clubs and choose the best goalkeeper based upon the following KPI's

 1. goalkeeping_diving
 2. goalkeeping_handling
 3. goalkeeping_kicking
 4. goalkeeping_reflexes

```python
# filtering data to find goalkeepers from other clubs except Real Madrid CF
df_gk = df[(df['player_positions'] == 'GK') & (df['club_name'] != 'Real Madrid CF') & (df['age'] <= 26)]
df_gk
```

```python
# importing necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
gk_diving = df_gk.sort_values('goalkeeping_diving', ascending = False)
plt.bar(gk_diving['short_name'].head(),gk_diving['goalkeeping_diving'].head())
```

```python
gk_handling = df_gk.sort_values('goalkeeping_handling', ascending = False)
plt.bar(gk_handling['short_name'].head(),gk_handling['goalkeeping_handling'].head())
```

```python
gk_kicking = df_gk.sort_values('goalkeeping_kicking', ascending = False)
plt.bar(gk_kicking['short_name'].head(),gk_kicking['goalkeeping_kicking'].head())
```

```python
gk_reflexes = df_gk.sort_values('goalkeeping_reflexes', ascending = False)
plt.bar(gk_reflexes['short_name'].head(),gk_reflexes['goalkeeping_reflexes'].head())
```

### Best Goalkeeper

G. Donnarumma has the best diving, handling and reflexes and is also the 4th best player in goal kicking. He is also young with age of 22 years, So I feel this player should be scouted for goalkeeping.
